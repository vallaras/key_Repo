 Linear regression,Ridge And Lasso

```python
#house pricing dataset
from sklearn.datasets import load_boston
```

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=load_boston()
```

```python
df
```

```python
pd.DataFrame(df.data)
```

```python
dataset=pd.DataFrame(df.data)
dataset.columns=df.feature_names
```

```python
dataset
```

```python
dataset['price']=df.target
```

```python
dataset.head()
```

```python
## Dividing the dataset into indepandent and dependent features
```

```python
X=dataset.iloc[:,:-1]##independent feature
Y=dataset.iloc[:,-1]## dependend feature
```

```python
Y.head()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(
X,Y, test_size=0.33, random_state=42)
```

```python
##linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
mse=cross_val_score(lin_reg,x_train,y_train,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()

params={'alpha':[1e-15,1e-10,1e-2,1,5,10,20]}
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,Y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
#Lasso Regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()

params={'alpha':[1e-15,1e-10,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=10)
lasso_regressor.fit(x_train,y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
from sklearn.model_selection import train_test_split
```
